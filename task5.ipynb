```python
#social media post engagement analyzer
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Using loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    # Condition check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    return total_likes, status


# Example Data
likes_data = [200, 300, 150, 400]   # Total = 1050

total, result = analyze_engagement(likes_data)

print("Total Likes:", total)
print("Post Status:", result)
```

```python
#rainfall adequacy checker
def rainfall_status(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status

# Example
rainfall_data = [70, 75, 65, 78]
required = 70

avg, result = rainfall_status(rainfall_data, required)
print("Average Rainfall:", avg)
print("Rainfall Status:", result)
```

```python
#duplicate account detection
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"

# Example
users = ["srikala", "ram", "john", "ram"]

print("Duplicate Accounts Found:", check_duplicates(users))
```

```python
#premium crop price
def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium

# Example
crop_prices = [1500, 2500, 1800, 3200]
print("Premium Crops:", premium_crops(crop_prices))
```

```python
#Application health checker
def app_health(errors):
    if errors == 0:
        return "Healthy"
    elif errors <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"

# Example
error_count = 7
print("Error Count:", error_count)
print("System Status:", app_health(error_count))
```

```python
#Student Attendance eligibility
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:   # 1 means present
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        result = "Eligible"
    else:
        result = "Not Eligible"

    return percentage, result

# Example
attendance = [1, 1, 1, 0, 1]
percent, status = attendance_eligibility(attendance)

print("Attendance Percentage:", percent)
print("Exam Eligibility:", status)
```

```python
#smart electricit bill analyzer
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

# Example
units = 250
bill, usage = calculate_bill(units)

print("Total Bill:", bill)
print("Usage Level:", usage)
```
