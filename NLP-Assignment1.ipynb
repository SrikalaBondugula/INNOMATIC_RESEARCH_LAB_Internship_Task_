```python
#task2

import re

def preprocess_text(text):
    if not text:
        return [], ""

    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)  # remove URLs/emails
    text = re.sub(r'\d+', '', text)                    # remove numbers
    text = re.sub(r'(.)\1+', r'\1\1', text)            # handle repeated chars
    text = text.lower()                               # lowercase
    text = re.sub(r'[^a-z\s]', '', text)              # remove special chars
    text = re.sub(r'\s+', ' ', text).strip()          # remove extra spaces

    tokens = text.split()

    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    clean_sentence = " ".join(tokens)

    return clean_sentence
preprocess_text("sooooooo goood")  
```

```python
#tas3
import re

def preprocess_text(text):
    if not text:
        return [], ""

    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1\1', text)
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence   # ✅ ONLY 2 values
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely loooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Noooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean= preprocess_text(sentence)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
    print("-" * 50)
```

```python
tokens = ["this", "is", "good", "good"]

def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(w) for w in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length


total, unique, avg = token_analytics(tokens)

print("Total:", total)
print("Unique:", unique)
print("Average Length:", avg)
```

```python
#task5
from collections import Counter

def frequency_analysis(all_tokens):
    counter = Counter(all_tokens)

    most_common = counter.most_common(10)
    least_common = counter.most_common()[:-6:-1]

    return most_common, least_common
tokens = ["good", "bad", "good", "happy", "bad", "good"]

most, least = frequency_analysis(tokens)

print("Most Common:", most)
print("Least Common:", least)
```

```python
#task 6
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
sentences = [
    "I love this product!!!",
    "Worst service ever",
    "Not happy with this"
]

result = full_pipeline(sentences)

print(result)
```

```python
#task7
# Edge case test inputs
edge_cases = [
    "",                         # Empty string
    "1234567890",               # Only numbers
    "😍😂🔥",                  # Only emojis
    "!!!@@@###",               # Only special characters
    "   ",                     # Only spaces
    None                       # None input
]

print("===== TASK 7: ERROR HANDLING =====\n")

for case in edge_cases:
    try:
        tokens, clean = preprocess_text(case)

        print("Input:", case)
        print("Tokens:", tokens)
        print("Clean Sentence:", clean)
        print("-" * 50)

    except Exception as e:
        print("Input:", case)
        print("Error:", str(e))
        print("-" * 50)
```
