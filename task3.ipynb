```python
#employee performance bonus eligibility
employee={
    "Ravi":92,
    "Anita":88,
    "kiran":92,
    "suresh":85
}
highest=0
for names,score in employee.items():
    if employee[names]>highest:
        highest=employee[names]
highest_list=[]
for names,score in employee.items():
    if employee[names]==highest:
        highest_list.append(names)
print(f"top Performers Eligible for Bonus:",end="")
if len(highest_list)==1:
    print(f"{highest_list[0]} (score:{highest})")
else:
    for i in range(len(highest_list)):
        if i==len(highest_list)-1:
             print(f"{highest_list[i]} (score:{highest})")
        else:
             print(f"{highest_list[i]}",end=", ")
   
```

```python
#Search Query Keyword Analysis
input_query=list(input().split())
print(input_query)
dic1={}
dic2={}
for i in input_query:
    if i in dic1:
        dic1[i]+=1
    else:
        dic1[i]=1
for word,value in dic1.items():
    if dic1[word]>1:
        dic2[word]=dic1[word]
print(dic2)
```

```python
#sensor data Validation
sensor_readings=list(map(int,input().split()))
even_list=[]
for i in range(len(sensor_readings)):
    if sensor_readings[i]%2==0:
        tupl=(i,sensor_readings[i])
        even_list.append(tupl)
print(f"valid sensor Readings(Hour,Value):{even_list}")
```

```python
#email domain usage analysis
emails=["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]
gmail_count=0
yahoo_count=0
for i in emails:
    if "gmail" in i:
        gmail_count+=1
    else:
        yahoo_count+=1
print(f"gmail.com:{int((gmail_count/len(emails))*100)}%")
print(f"yahoo.com:{int((yahoo_count/len(emails))*100)}%")
```

```python
#sale spike detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
total = 0
for value in sales:
    total += value

average = total / len(sales)
threshold = average + (0.3 * average)
for i in range(len(sales)):
    if sales[i] > threshold:
        print(f"Day {i+1}: {sales[i]}")
```

```python
#duplicate user id detection
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
frequency = {}
for user in user_ids:
    if user in frequency:
        frequency[user] = frequency[user] + 1
    else:
        frequency[user] = 1
for user in frequency:
    if frequency[user] > 1:
        print(user, "->", frequency[user], "times")
```
