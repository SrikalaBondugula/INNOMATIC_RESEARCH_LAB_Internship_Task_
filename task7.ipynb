```python
#smart light controller
class light_controller:
    def __init__(self,name):
        self.name=name
        self.status="off"
    def action(self,act):
        if act=="on":
            self.status="ON"
            print(f"{self.name} is {self.status}")
        elif act=="off":
            print(f"{self.name} is {self.status}")
light_name=input("enter").lower()
action=input("enter").lower()
l1=light_controller(light_name)
l1.action(action)
```

```python
#employee id card
class id_card_system:
    def __init__(self,e_name,id,dept):
        self.e_name=e_name
        self.id=id
        self.dept=dept
    def display(self):
        print("Employee ID Card")
        print(f"Employee Name:{self.e_name}")
        print(f"Name:{self.id}")
        print(f"Department:{self.dept}")
E_name=input("enter employee name")
Id=input("enter employee id")
Dept=input("enter department")
ics1=id_card_system(E_name,Id,Dept)
ics1.display()
```

```python
#mobile contact record
class Contact_record:
    def __init__(self,contact_name,phone_number):
        self.contact_name=contact_name
        self.phone_number=phone_number
    def display(self):
        print("Contact saved")
        print(f"Name:{self.contact_name}")
        print(f"Phone:{self.phone_number}")
name=input("enter name")
Contact=int(input("enter the phone number"))
c1=Contact_record(name,Contact)
c1.display()
```

```python
#Product Price Tag Generator
class Tag_generator:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def display(self):
        print(f"product:{self.name}")
        print(f"Price:{self.price}")
product_name=input("enter")
product_price=int(input("enter price"))
T1=Tag_generator(product_name,product_price)
T1.display()
```

```python
#Movie Rating display system
class movie_rating:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
        self.total_rating=5
    def display(self):
        print(f"movie:{self.name}")
        print(f"Rating:{self.rating}/{self.total_rating}")
moviename=input("enter the movie name")
Ratings=float(input("enter the ratings"))
m1=movie_rating(moviename,Ratings)
m1.display()

```
