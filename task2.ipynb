```python
#Unique Words in a sentence
sentence=input("enter any sentence")
list1=list(sentence.split(" "))
set1=set( )
for words in list1:
    set1.add(words)
print(f"Unique words Count:{len(set1)}")
print(f"Unique words:{set1}")
```

```python
#Highest salary from employee data
employee={
    "Ravi":75000,
    "Anita":68000,
    "kiran":72000
}
highest_salary=0
employee_name=""
for name,salary in employee.items():
    if salary>highest_salary:
        highest_salary=salary
        employee_name=name
print(f"highest salary:{employee_name}-{highest_salary}")
```

```python
#find maximum and minimum
input_list=list(map(int,input().split()))
minimum=input_list[0]
maximum=input_list[0]
for value in input_list:
    if value>maximum:
        maximum=value
    if value<minimum:
        minimum=value
print(input_list)
print(f"max:{maximum}")
print(f"min:{minimum}")

```

```python
#count products above a price threshold
products_list=list(map(int,input().split()))
count=0
threshold=1000
for price in products_list:
    if price>threshold:
        count+=1
print(f"products above {threshold}:{count}")
```

```python
#calculate attendance percentage
attendance_list=list(input("enter the attendance").split())
present_count=0
for i in attendance_list:
    if i=="p":
        present_count+=1
print(f"Attendance Percentage:{(present_count/len(attendance_list))*100}")
```

```python
#remove duplicate phone numbers
phonenum_list=list(map(int,input().split()))
unique_set=set( )
for num in phonenum_list:
    unique_set.add(num)
print(f"Unique phone numbers:{unique_set}")

```

```python
#count character Frequency
input_string=input("enter the string")
string_dict={}
for char in input_string:
    if char in string_dict:
        string_dict[char]+=1
    else:
        string_dict[char]=1
print(string_dict)
```
