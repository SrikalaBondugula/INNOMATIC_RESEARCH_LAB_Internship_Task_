```python
#Smart parking Lot Management system
def smartparking(logs):
    parking_capacity=50
    parked_vehicles=0
    parking_status="Available"
    for i in logs:
        if i=="IN":
            parked_vehicles+=1
    if parking_capacity-parked_vehicles==0:
        parking_status="Not available"
    return parking_status,parked_vehicles
total_num_vehicles=int(input("enter number of vehicles"))
list1=[]
for i in range(total_num_vehicles):
    status=input("enter status of the vehicle").upper()
    list1.append(status)
parked_status,parked_vehicles=smartparking(list1)
print(f"currently parked vehicles:{parked_vehicles}")
print("parking status :",parked_status)



    
```

```python
#online food delivery time estimator
def estimator(distance,weather,trafficlevel):
    speed=30
    time=distance/speed*60
    if trafficlevel=="medium":
        time+=10
    elif trafficlevel=="high":
        time+=20
    if weather=="rainy":
        time+=20
    return time
Distance=int(input("enter the distance"))
Weather=input("enter the weather condition").lower()
Trafficlevel=input("enter the traffic level").lower()
Time=estimator(Distance,Weather,Trafficlevel)
print("Estimated Delivery Time:",Time,"minutes")
```

```python
#movie thetre seat occupancy
def theatre(booked_seats_list):
    total_seats=200
    booked=0
    status="empty"
    for i in booked_seats_list:
        if i==1:
            booked+=1
    remaining=total_seats-booked
    occupancy=(booked/total_seats)*100
    if remaining==0:
        status="full"
    elif remaining<50:
        status="almost full"
    elif remaining<100:
        status="Available"
    return occupancy,status
seats_list=list(map(int,input("enter").split()))
Occupancy,Status=theatre(seats_list)
print("occupancy:",int(Occupancy),"%")
print("show status:",status)


```

```python
#cloud server load classification system
def serverload(cpu_readings):
    sum=0
    for i in cpu_readings:
        sum+=i
    avg=sum//len(cpu_readings)
    if avg>=80:
        server_status="critical"
    elif avg>=50:
        server_status="warning"
    else:
        server_status="Normal"
    return avg,server_status
cpu_list=list(map(int,input().split()))
Avg,Server_status=serverload(cpu_list)
print("Average CPU Load:",Avg,"%")
print("Server Status:",Server_status)
```

```python
#smart class room resource usage monitor
def resource_usage_monitor(resources):
    max_usage=8
    energy_alert="No"
    for key,value in resources.items():
        if resources[key]>=max_usage:
            print("overused resource:",key)
            energy_alert="yes"
            print("Energy Alert:",energy_alert)
n_of_resources=int(input("enter the number of resources"))
dic1={}
for i in range(n_of_resources):
    key=input("enter the key")
    value=int(input("enter the value for key"))
    dic1[key]=value
resource_usage_monitor(dic1)
```

```python
#Online event registration capacity controller
def Registration(registrations):
    Event_capacity=100
    registration_status="open"
    wait_listed_users=0
    confirmed_registrations=0
    if registrations>=Event_capacity:
        wait_listed_users=registrations-Event_capacity
        registration_status="closed"
        confirmed_registrations=Event_capacity
    else:
        confirmed_registrations=registrations
    return confirmed_registrations,wait_listed_users,registration_status
num_registrations=int(input("enter the number of registrations"))
Confirmed_registrations,Waitlisted_users,Registrations_status=Registration(num_registrations)
print("Confirmed Registrations:",Confirmed_registrations)
print("Waitedlisted Users:",Waitlisted_users)
print("Registrations Status:",Registrations_status)
    
```
