```python
#ERROR MESSAGE DETECTOR
logs=["INFO","ERROR","WARNING","ERROR"]
count=0
for i in logs:
    if i=="ERROR":
        count+=1
print(f"Total ERROR count={count}")
```

```python
#MESSAGE LENGTH ANALYZER
messages=["Hi","Welcome to the Platform","ok"]
for string in messages:
    length=0
    for j in string:
        length+=1
    if length>10:
        print(f"message:{string}")
        print(f"length:{length}")
        print("Flag: greater than 10")
    else:
         print(f"message:{string}")
         print(f"length:{length}")

```

```python
#SIMPLE DATA CLEANER
names=[" Alice ","bob","CHARLIE "]
new_list=[]
for i in names:
   list1=list(i)
   for j in list1:
      if j==" ":
         list1.remove(j)
   string="".join(list1)
   new_list.append(string.lower())

print(new_list)

```

```python
#PASS/FAIL ANALYZER
marks=[45,78,90,33,60]
total_pass_stds=0
total_fail_stds=0
for i in marks:
    if i>=50:
        total_pass_stds+=1
    else:
        total_fail_stds+=1
print(f"Total number of passed students: {total_pass_stds}")
print(f"Total number of failed students: {total_fail_stds}")
```
